```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('xg.csv')
```

```python
df
```

```python
a_xG = [0]
h_xG= [0]
a_min = [0]
h_min = [0]

hteam = df['team'].iloc[0]
ateam = df['team'].iloc[-1]

for x in range(len(df['xG'])):
    if df['team'][x]==ateam:
        a_xG.append(df['xG'][x])
        a_min.append(df['minute'][x])
    if df['team'][x]==hteam:
        h_xG.append(df['xG'][x])
        h_min.append(df['minute'][x])
        
def nums_cumulative_sum(nums_list):
    return [sum(nums_list[:i+1]) for i in range(len(nums_list))]

a_cumulative = nums_cumulative_sum(a_xG)
h_cumulative = nums_cumulative_sum(h_xG)
```

```python
fig, ax = plt.subplots(figsize = (10,5))
fig.set_facecolor('Black')
ax.patch.set_facecolor('Black')

plt.xticks([0,15,30,45,60,75,90])
plt.yticks([0.0,0.5,1.0,1.5,2.0,2.5,3.0])
plt.xlabel('Minute',fontname='Calibri',color='white',fontsize=16)
plt.ylabel('xG',fontname='Calibri',color='white',fontsize=16)


ax.step(x=a_min,y=a_cumulative,color='Blue',label=ateam,linewidth=3,)
ax.step(x=h_min,y=h_cumulative,color='Red',label=hteam,linewidth=3)

for spine in ['left', 'right', 'top', 'bottom']:
    ax.spines[spine].set_color('white')  
    ax.spines[spine].set_linewidth(1)     

ax.legend()

image_path = 'south.png'  
image = plt.imread(image_path)
image_extent = [0.15, 0.9, 0.7, 0.4]  
axins = fig.add_axes(image_extent)
axins.imshow(image, alpha=1) 
axins.axis('off') 
```
